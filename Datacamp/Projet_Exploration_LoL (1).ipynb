### Avec le displot nous ne pouvons pas trop conclure, ni bien voir l'évolution des champions en fonction du pickrate et du winrate

```python
sns.pairplot(data_championV2,hue='Role')
```

### Le pairplot ici est déjà mieux, nous nous baserons donc sur des graphiques de ce type durant la projet.

```python
g = sns.PairGrid(data_championV2, hue="Class", corner=True)
g.map_lower(sns.kdeplot, hue=None, levels=5, color=".2")
g.map_lower(sns.scatterplot, marker="+")
g.map_diag(sns.histplot, element="step", linewidth=0, kde=True)
g.add_legend(frameon=True)
g.legend.set_bbox_to_anchor((.61, .6))
```

### Avec le Pairgrid aussi nous pouvons voir les regroupements des différents champions en fonction de leur caractéristique.

```python
fig = sns.FacetGrid(data_championV2, col="Role", hue="Class", aspect=.75, height=4)
fig.map_dataframe(sns.scatterplot, x="Winrate", y="Pickrate")  # bill_depth_mm vs body_mass_g in the 3 islands and the 3 species
fig.set_axis_labels("Winrate (%)", "Pickrate (%)")
fig.add_legend()
```

## Analyse de la Toplane

```python
data_championTOPV2 = data_championTOPV2.rename(columns= {'Win %' : 'Winrate', 'Role %' : 'Rolerate'})
```

```python
data_championTOPV2 = data_championTOPV2.sort_values(by=['Winrate'],ascending=False)
data_championTOPV2
```

```python
data_championTOPV2['Winrate'] = data_championTOPV2['Winrate'].str[:-1]
data_championTOPV2['Winrate'] = data_championTOPV2['Winrate'].astype('float')
```

```python
data_championTOPV2['Rolerate'] = data_championTOPV2['Rolerate'].str[:-1]
data_championTOPV2['Rolerate'] = data_championTOPV2['Rolerate'].astype('float')
```

```python
sns.jointplot(data = data_championTOPV2, x = "Winrate", y = "Score", hue = "Class")
```

### Le score en fonction du Winrate ne veux rien dire, car un personnage joué une ou deux fois avec un totalité de victoire aura un Winrate de 100% et donc un score maximum, ce qui n'est pas représentatif et fausse les données.

```python
data_championTOPV2 = data_championTOPV2.rename(columns= {'Pick %' : 'Pickrate'})
data_championTOPV2['Pickrate'] = data_championTOPV2['Pickrate'].str[:-1]
data_championTOPV2['Pickrate'] = data_championTOPV2['Pickrate'].astype('float')
```

```python
sns.jointplot(data = data_championTOPV2, x = "Pickrate", y = "Score", hue = "Class")
sns.jointplot(data = data_championTOPV2, x = "Pickrate", y = "Winrate", hue = "Class")
```

### Ici le Score en fonction du Pickrate (1er graphique) est déjà plus représentatif car plus un personnage est joué et plus son Score sera révélateur de sa "force" dans la méta et du patch. 
#### Ex : Aatrox est l'un des personnages les plus joués et a un bon Score car le rapport entre Pickrate et le Winrate lui attribu se Tier God qui veux bien dire que c'est actuellement le meilleur Toplaner du patch.

### Ensuite le Winrate en fonction du Pickrate (2ème graphique) viens appuyer le 1er graphique et montre bien son Tier.
#### Aatrox étant l'un des champions les plus "pick", même si son Winrate est en dessous d'autres champions, le Score se base sur le rapport entre le fait qu'il ait beaucoup été joué et son taux de match gagnés.

```python
fig = px.bar(data_championTOPV2,x="Winrate", y="Pickrate",hover_data = ['Name'])
fig.update_traces(width=0.2)
fig.show()
```

### Le graphique ci-dessus permettra d'avoir le nom du champion plus facilement.

```python
sns.histplot(data = data_championTOPV2, x = "Winrate", y = "Pickrate", hue = "Class")
```
